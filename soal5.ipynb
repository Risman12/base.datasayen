```python
print('==================== soal 2 ====================')
print()
pajak = 0.03

utang1 = 0.08
utang2 = 0.04
utang3 = 0.12
utang4 = 0.11
utang5 = 0.03

laba1 = 32500000
laba2 = 31000000
laba3 = 42000000
laba4 = 51500000
laba5 = 27000000

juml =(pajak * laba1) + (utang1 * 280000000)
print("total pajak yang di keluar kan  perusahaan A : " + str (juml))
juml =(pajak * laba2) + (utang2 * 310000000)
print("total pajak yang di keluar kan  perusahaan B : " + str (juml))
juml =(pajak * laba3) + (utang3 * 420000000)
print("total pajak yang di keluar kan  perusahaan C : " + str (juml))
juml =(pajak * laba4) + (utang4 * 550000000)
print("total pajak yang di keluar kan  perusahaan D : " + str (juml))
juml =(pajak * laba5) + (utang5 * 340000000)
print("total pajak yang di keluar kan  perusahaan E : " + str (juml))
print()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
eskul = ['A','B','C','D','E']
pengikut = [23375000, 13330000, 51660000, 62045000, 11010000]
c = ['red', 'blue', 'green']
plt.title("total semua pajak ")
plt.bar(eskul, pengikut, color = c)
plt.show()
print()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
jurusan = ['A','B','C','D','E']
siswa = [23375000, 13330000, 51660000, 62045000, 11010000]
ax.pie(siswa,labels=jurusan,autopct= '%1.2f%%')
plt.title("JUMLAH SEMUA PAJAK ")
plt.show()
```
