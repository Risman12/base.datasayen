```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
nama = ['tia','hana','widia','romi','samsul','dimas']
jumlah_belanja = [246600, 193200, 119300, 221700, 323800, 379300]
c = ['red', 'blue', 'green']
plt.title("Jumlah belanjaan")
plt.bar(nama, jumlah_belanja, color = c)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
nama = ['tia','hana','widia','romi','samsul','dimas']
jumlah_belanja = [246600, 193200, 119300, 221700, 323800, 379300]
ax.pie(jumlah_belanja,labels=nama,autopct= '%1.2f%%')
plt.title("jumlah belanja belanja")
plt.show()
```
